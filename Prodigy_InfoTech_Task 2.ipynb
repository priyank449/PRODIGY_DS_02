# Task 2 :Perform data cleaning and exploratory data analysis (EDA) on a Titanic dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("titanic.csv",skipinitialspace = True)
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
print(data.isnull().sum())
```

```python
data['Cabin'].fillna(data['Cabin'].mode()[0], inplace = True)
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace = True)
data['Age'].fillna(data['Age'].mean(), inplace = True)
```

```python
print(data.isnull().sum())
```

```python
data.head()
```

```python
sns.boxplot(x = 'Age',y = 'Sex', data = data)
plt.title("Age Distribution Genderwise", fontsize = 20)
plt.xlabel("Age",fontsize = 15)
plt.ylabel("Sex",fontsize = 15)
```

```python
clas = data.groupby('Pclass')['Survived'].mean()
print(clas)
```

```python
plt.bar(clas.index, clas.values)
plt.xlabel('Pclass',fontsize = 15)
plt.ylabel('Survival Rate',fontsize = 15)
plt.title('Survival Rate by Pclass',fontsize = 20)
plt.show()
```
